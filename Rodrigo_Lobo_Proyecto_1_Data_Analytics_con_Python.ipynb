##	 Cual es la categoría con mayor crecimiento?

```python
big_db['Category'].nunique()
```

```python
big_db['Category'].unique()
```

```python
category_n_profit = big_db[['Order Date','Category','Profit']]
category_n_profit.head(15)
```

```python
category_n_profit.rename(columns = {'Order Date':'Order_Date'}, inplace = True)
category_n_profit.info()
category_n_profit.tail(5)
```

```python
category_n_profit['Order_Date'] = pd.to_datetime(category_n_profit['Order_Date'], format= '%d-%m-%Y')
category_n_profit['Year']= category_n_profit['Order_Date'].dt.year
category_n_profit['Month']= category_n_profit['Order_Date'].dt.month
category_n_profit['Day']= category_n_profit['Order_Date'].dt.day
category_n_profit.info()
```

```python


cat_n_pro = category_n_profit.drop(columns='Order_Date')
cat_n_pro.head()
```

```python
cat_n_pro_final = cat_n_pro.groupby(['Category', 'Year'])['Profit'].sum().reset_index()
cat_n_pro_final
```

```python

sns.lineplot(data=cat_n_pro_final, x='Year', y='Profit', ci=None, hue= 'Category',palette=['red','green','yellow'])
```

como podemos observar de la frafica la categoria con mayor crecimiento sobre el tiempo es la categoria de tecnologia

```python
tec_best_sel = big_db[['Category','Sub-Category']]
tempdf = tec_best_sel[tec_best_sel['Category'] == 'Technology']
tempdf.groupby(['Sub-Category']).count()
```

dentro de la categoria de tecnologia lo que mas se vende son telefonos

Cual es la Region con mejores ventas?

```python
region_n_profit = big_db[['Region','Profit','Shipping Cost','Sales']]
region_n_profit
```

```python
region_n_profit.rename(columns = {'Shipping Cost':'Shipping_Cost'}, inplace = True)
sum_pro_reg = region_n_profit.groupby(['Region'])['Profit'].sum().reset_index()
sum_disc_reg = region_n_profit.groupby(['Region'])['Shipping_Cost'].sum().reset_index()
sum_sale_reg = region_n_profit.groupby(['Region'])['Sales'].sum().reset_index()
region_temp= pd.merge(sum_sale_reg, sum_disc_reg, how ='outer', on ='Region')
region_final= pd.merge(region_temp, sum_pro_reg, how ='outer', on ='Region')
region_final
```

```python
fig, ax = plt.subplots(figsize=(15, 8), dpi=100)
sns.barplot(data=region_final, x="Region", y="Sales", hue="Region", width=0.8,dodge=False )
```

la grafica nos indica que la region que tien mayore ventas es la region central

```python
region_bestsel = big_db[['Region','Sub-Category']]
tempdf1 = region_bestsel[region_bestsel['Region'] == 'Central']
tempdf1.groupby(['Sub-Category']).count()
```

lo que mas vendemos en la region central son lo cartapasios y el producto menos vendido son las mesas

Relaciones entre costos de envio, ganancias, ventas y descuentos

```python
heatmap_df= big_db[['Shipping Cost','Profit', 'Sales','Discount']]
corr = heatmap_df.corr()
plt.figure(figsize=(12, 10))

sns.heatmap(corr, 
            cmap='flare', vmax=1.0, vmin=0, linewidths=0.1,
            annot=True, annot_kws={"size": 8}, square=True);
```

la grafica nos permite ver las relaciones entre la variables numericas que tenemos y se nos presenta que las variables numericas que muestran una relacion mas cercana son las ventas y el costo de envio


```python
discu_region = big_db[['Region','Discount']]
fig, ax = plt.subplots(figsize=(15, 8), dpi=100)
sns.barplot(data=discu_region, x="Region", y="Discount", hue="Region", width=0.8,dodge=False )
```

podemos ver que la mayor cantidad de descuentos esta en la region Southeast Asia que es un region que no tien un gran cantidad de ventas
